# p4p basics: write a PV

This is the simplest write example from the training slides.
It moves the simulated horizontal mirror and then reads the value back.

```python
from p4p.client.thread import Context
import time

ctx = Context('pva')

# Move horizontal mirror to position 10.0
ctx.put('LAB:SIM:HMIR', 10.0)
print('Motor moving...')

# Wait and read back
time.sleep(2)
pos = ctx.get('LAB:SIM:HMIR')
print(f'HMIR arrived at: {pos}')
```
