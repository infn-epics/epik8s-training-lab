# p4p basics: read a PV

This is the simplest read example from the training slides.
Run the code cell below to read one motor PV and the camera centroid.

```python
from p4p.client.thread import Context

# Create a PV Access context (connects to the EPICS network)
ctx = Context('pva')

# Read a motor position
pos = ctx.get('LAB:SIM:HMIR')
print(f'HMIR position: {pos}')

# Read camera centroid
cx = ctx.get('LAB:SIM:CAM01:Stats1:CentroidX_RBV')
cy = ctx.get('LAB:SIM:CAM01:Stats1:CentroidY_RBV')
print(f'Beam centroid: ({cx}, {cy})')
```
